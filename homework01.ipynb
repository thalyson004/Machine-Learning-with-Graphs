## Import librarys

```python
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns
```

## Custom functions

```python
def graph_by_list(data: list, xlabel: str = None, ylabel: str = None):
    sns.displot(data, kind="hist", height=4, aspect=2.5).set(
        xlabel=xlabel,
        ylabel=ylabel,
    )
```

## Get karate club network as G

```python
G = nx.karate_club_graph()
```

```python
nx.draw_circular(G, with_labels=True)
```

## Answers

### 1° 

Para cada uma das features {degree, eigenvector centrality, betweeness centrality , closeness centrality e clustering coefficient} calcular o valor da feature para cada vértice e exibir um gráfico de distribuição (um gráfico para cada feature, onde a variável considerada é a feature calculada em cada vértice).Para cada uma das features {degree, eigenvector centrality, betweeness centrality , closeness centrality e clustering coefficient} calcular o valor da feature para cada vértice e exibir um gráfico de distribuição (um gráfico para cada feature, onde a variável considerada é a feature calculada em cada vértice).

```python
# Degree
degrees = [value for node, value in nx.degree(G)]
graph_by_list(degrees, xlabel="Degrees", ylabel="Count")
```

```python
# Eigenvector centrality
eigen = nx.eigenvector_centrality(G)
graph_by_list(eigen, xlabel="Eigenvector Centrality", ylabel="Count")
```

```python
# Betweeness centrality
betweeness = nx.betweenness_centrality(G)
graph_by_list(betweeness, xlabel="Betweeness Centrality", ylabel="Count")
```

```python
# Closeness centrality
closeness = nx.closeness_centrality(G)
graph_by_list(closeness, xlabel="Closeness Centrality", ylabel="Count")
```

```python
# Clustering coefficient
clustering = nx.clustering(G)
graph_by_list(clustering, xlabel="Clustering Coefficient", ylabel="Count")
```

### 2°

Para cada uma das features {distance-based, jaccard, adamic-adar, e Katz index} calcular o valor da feature para cada par de vértices e exibir o resultado na forma de um gráfico de distribuição (um gráfico para para cada feature, onde a variável considerada é a feature calculada para cada par de vértices).

```python
# Distance-based
distances = []
for index, features in G.nodes(data=True):
    distances.extend(list(dict(nx.single_source_shortest_path_length(G, index)).values()))

graph_by_list(distances, xlabel="Distance", ylabel="Count")
```

```python
# Jaccard
jaccard = [value for (u, v, value) in list(nx.jaccard_coefficient(G))]

graph_by_list(jaccard, xlabel="Jaccard", ylabel="Count")
```

```python
# Adamic-adar
adamic = [value for (u, v, value) in list(nx.adamic_adar_index(G))]

graph_by_list(adamic, xlabel="Adamic-adar", ylabel="Count")
```
